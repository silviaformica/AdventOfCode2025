```python
data = '''.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............'''

data = data.split('\n')
```

```python
file_path = './inputDay7.txt'

with open(file_path, 'r') as file:
    data = file.read()

data = data.split('\n')
# the last item is nothing, we drop it
data = data[:-1]
```

```python
m = np.empty([len(data), len(data[0])], dtype='<U5')
```

```python
for s, string in enumerate(data):
    m[s, :] = np.array(list(string))
```

```python
counter = 0

for r in range(m.shape[0]-1):

    for c in range(m.shape[1]-1):

        if m[r, c] == 'S':
            m[r, c] = '|'

        if m[r, c] == '|' and m[r+1, c] == '.':
            m[r+1, c] = '|'

        if m[r, c] == '|' and m[r+1, c] == '^':
            m[r+1, c-1] = '|'
            m[r+1, c+1] = '|'
            counter +=1

print(f"Splitted a total of {counter} times")
```
