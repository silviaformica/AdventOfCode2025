```python
# loading module
import numpy as np

# loding input ranges
ranges = "4487-9581,755745207-755766099,954895848-955063124,4358832-4497315,15-47,1-12,9198808-9258771,657981-762275,6256098346-6256303872,142-282,13092529-13179528,96201296-96341879,19767340-19916378,2809036-2830862,335850-499986,172437-315144,764434-793133,910543-1082670,2142179-2279203,6649545-6713098,6464587849-6464677024,858399-904491,1328-4021,72798-159206,89777719-90005812,91891792-91938279,314-963,48-130,527903-594370,24240-60212"
ranges = ranges.split(',')
```

```python
import numpy as np

invalidID = []
counter = 0
for i, thisrange in enumerate(ranges):
    start = int(thisrange.split('-')[0])
    stop = int(thisrange.split('-')[1])
    for ii, ID in enumerate(np.arange(start, stop + 1)):
        ID_s = str(ID)
        split = int(len(ID_s)/2)
        firsthalf = ID_s[:split]
        secondhalf = ID_s[split:]
        if firsthalf == secondhalf:
            counter += 1
            invalidID.append(ID)

print(f"Number of invalid IDs: {counter}")
print(f"Sum of all invalid IDs: {np.asarray(invalidID).sum()}")
```

## Part 2

```python
import math
import textwrap

invalidID = []

for i, thisrange in enumerate(ranges):
    start = int(thisrange.split('-')[0])
    stop = int(thisrange.split('-')[1])
    # looping through all IDs
    for ii, ID in enumerate(np.arange(start, stop + 1)):
        ID_s = str(ID)
        # splitting for all divisors of the length of ID
        for i in range(1, int(len(ID_s))):
            if len(ID_s) % i == 0:       # If 'i' is a divisor of 'ID'
                splits = textwrap.wrap(ID_s, i)
                # check if it is the first element repeated for the whole list
                if splits.count(splits[0]) == len(splits):
                    invalidID.append(ID)

# removing duplicates
invalidID = list(set(invalidID))

print(f"Number of invalid IDs: {len(invalidID)}")
print(f"Sum of all invalid IDs: {np.asarray(invalidID).sum()}")
```
