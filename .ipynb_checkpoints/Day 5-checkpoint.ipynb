## Part 1

```python
import numpy as np
```

```python
data = '''3-5
10-14
16-20
12-18

1
5
8
11
17
32'''


ranges = data.split('\n\n')[0]
ids = data.split('\n\n')[1]

ranges = ranges.split('\n')
ids = ids.split('\n')
ids = [int(x) for x in ids]

fresh = []
for r in ranges:
    fresh.extend(np.arange(int(r.split('-')[0]), int(r.split('-')[1])+1))

print(f"Fresh items: {list(set(fresh).intersection(ids))}")

print(f"Total number of fresh items: {len(list(set(fresh).intersection(ids)))}")
```

This basic approach does not work for the full input because numbers are too large and cause memory issues.
So I am switching approach.

```python
file_path = './inputDay5.txt'

with open(file_path, 'r') as file:
    data = file.read()

ranges = data.split('\n\n')[0]
ids = data.split('\n\n')[1]

ranges = ranges.split('\n')
ids = ids.split('\n')

ids = [int(x) for x in ids if x != '']
```

```python
fresh = []
for i in ids:
    for r in ranges:
        start_range = int(r.split('-')[0])
        end_range = int(r.split('-')[1])
        if i >= start_range and i<= end_range:
            fresh.append(i)

print(f"Total number of fresh items: {len(np.unique(np.asarray(fresh)))}")
```
