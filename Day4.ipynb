# Advent of Code - Day 4

## Part 1

```python
import pandas as pd
import numpy as np
```

```python
text = np.loadtxt('./input.txt', dtype = 'str')
data = np.zeros([text.shape[0], len(text[0])])

for i, row in enumerate(text):
    row_wrapped = list(row)
    data[i, :] = data[i, :] = [1 if x == '@' else 0 for x in row_wrapped]
```

```python
counts = np.zeros(data.shape)
targets = np.zeros(data.shape)
for row_idx in range(data.shape[0]):
    for col_idx in range(data.shape[0]):

        start_row = row_idx-1
        end_row = row_idx+2
        start_col = col_idx-1
        end_col = col_idx+2
        
        if row_idx == 0 :
            start_row = row_idx
        elif row_idx >= data.shape[0]-1:
            end_row = row_idx+1

        if col_idx == 0 :
            start_col = col_idx
        elif col_idx >= data.shape[1]-1:
            end_col = col_idx+1
            
        subset = data[start_row:end_row, start_col:end_col]
        
        if np.sum(subset.shape[0] + subset.shape[1])< 4:
            print('issue!' + str(row_idx) + ', ' + str(col_idx))
        elif np.sum(subset.shape[0] + subset.shape[1])> 9:
            print('issue!' + str(row_idx) + ', ' + str(col_idx))
        
        counts[row_idx, col_idx] = np.sum(subset)

        
        if counts[row_idx, col_idx] <= 4 and data[row_idx, col_idx] == 1:
            targets[row_idx, col_idx] = 1

print(f"Final number of accessible items: {int(np.sum(targets))}")
```

## Part 2

```python
# reset 


total_count = 0
this_round  = 0

while np.sum(targets) != 0:

    counts = np.zeros(data.shape)
    targets = np.zeros(data.shape)
    for row_idx in range(data.shape[0]):
        for col_idx in range(data.shape[0]):
    
            start_row = row_idx-1
            end_row = row_idx+2
            start_col = col_idx-1
            end_col = col_idx+2
            
            if row_idx == 0 :
                start_row = row_idx
            elif row_idx >= data.shape[0]-1:
                end_row = row_idx+1
    
            if col_idx == 0 :
                start_col = col_idx
            elif col_idx >= data.shape[1]-1:
                end_col = col_idx+1
                
            subset = data[start_row:end_row, start_col:end_col]
            
            if np.sum(subset.shape[0] + subset.shape[1])< 4:
                print('issue!' + str(row_idx) + ', ' + str(col_idx))
            elif np.sum(subset.shape[0] + subset.shape[1])> 9:
                print('issue!' + str(row_idx) + ', ' + str(col_idx))
            
            counts[row_idx, col_idx] = np.sum(subset)
    
            
            if counts[row_idx, col_idx] <= 4 and data[row_idx, col_idx] == 1:
                targets[row_idx, col_idx] = 1

    print(f"Rolls removed in round {this_round}: {int(np.sum(targets))}") 
    this_round += 1

    data = data - targets

    total_count += int(np.sum(targets))

print(f"Final number of accessible items: {total_count}")
```
