# Advent of Code - Day 4

## Part 1

```python
import pandas as pd
import numpy as np
import textwrap
```

```python
text = np.loadtxt('./input.txt', dtype = 'str')
data = np.zeros([text.shape[0], len(text[0])])

for i, row in enumerate(text):
    row_wrapped = textwrap.wrap(row, 1)
    row_wrapped = row_wrapped[0].replace('.', '0').replace('@', '1')
    data[i, :] = row_wrapped
```

```python
counts = np.zeros(data.shape)
targets = np.zeros(data.shape)
for row_idx in range(data.shape[0]):
    for col_idx in range(data.shape[0]):

        start_row = row_idx-1
        end_row = row_idx+2
        start_col = col_idx-1
        end_col = col_idx+2
        
        if row_idx == 0 :
            start_row = row_idx
        elif row_idx >= data.shape[0]-2:
            end_row = row_idx

        if col_idx == 0 :
            start_col = col_idx
        elif col_idx >= data.shape[1]-2:
            col_col = col_idx
            
        subset = data[start_row:end_row, start_col:end_col]
        
        counts[row_idx, col_idx] = np.sum(subset)

        
        if counts[row_idx, col_idx] < 4 and data[row_idx, col_idx] == 1:
            targets[row_idx, col_idx] = 1

print(f"Final number of accessible items: {int(np.sum(targets))}")
```
