## Part 1

```python
import numpy as np
import pandas as pd
```

```python
file_path = './inputDay6.txt'

with open(file_path, 'r') as file:
    data = file.read()

data = data.replace(' ', ',').split('\n')

# the last item is nothing, we drop it
data = data[:-1]
```

```python
items = []
for i, line in enumerate(data):

    if i < len(data)-1:
        a = [int(x) for x in line.replace(' ', ',').split(',') if x != '']
        items.append(a)
    else:
        operations = [x for x in line.replace(' ', ',').split(',') if x != '']
    
```

```python
data = np.stack(items)
```

```python
total = 0
for i in range(data.shape[1]):
    if operations[i] == '+':
        result = np.sum(data[:, i])
    elif operations[i] == '*':
        result = np.prod(data[:, i])
    total += result

print(f"Final result: {total}")  
```

## Part 2

```python
file_path = './inputDay6.txt'

with open(file_path, 'r') as file:
    a = [list(line) for line in file]

data = np.empty([4, len(a[0])], dtype='<U5')

for i, row in enumerate(a[:-1]):
    data[i, :] = row
```

```python
# dropping last line which is just \n
data = data[:, :-1]
```

```python
allnums = []
for c in range(data.shape[1]):
    num = ''
    for pos in range(data.shape[0]):
        if data[pos, c]!= ' ':
            num = num + data[pos, c]
    allnums.append(num)
```

```python
from itertools import groupby

def split_by_value_groupby(lst, delimiter):
  # Using groupby to group consecutive elements 
  #based on whether they are equal to the delimiter
    return [list(group) for k, group in groupby(lst, lambda x: x == delimiter) if not k]

allnums = split_by_value_groupby(allnums, '')
len(allnums)
```

```python
data[:, -2:], allnums[-1]
```

```python
total = 0
for i, problem in enumerate(allnums):

    values = np.array([int(x) for x in problem])

    if operations[i] == '+':
        result = np.sum(values)
    elif operations[i] == '*':
        result = np.prod(values)
    total += result
    
print(f"Final result: {total}")      
```
